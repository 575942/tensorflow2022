```python
import tensorflow.keras
import numpy as np
```

# 构建模型

```python
#构建模型
model=tensorflow.keras.Sequential([
    tensorflow.keras.layers.Dense(units=1,
                                    input_shape=[1]  #输入值只有一个
                                    )  #一层神经网络，
])
#定义loss fuction和optimizer
model.compile(optimizer="sgd",  #定义优化器
                loss="mean_squared_error"  #定义损失，平方差
                )

#构建数据
x=np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
y=np.array([-3.0,1.0,1.0,3.0,4.0,8.0],dtype=float)
#训练模型
model.fit(x=x,y=y,epochs=600)  #epochs一个数据重复训练500次
```

# 模型预测

```python
print(model.predict([10.0]))
```
